```python
# 1.	 Vector-Based Distance Measures: Used mainly for numerical data.
# Euclidean Distance

import numpy as np
from scipy.spatial import distance

pointA = np.array([2, 4, 6])
pointB = np.array([5, 1, 9])

# Euclidean Distance
euclidean_dist = distance.euclidean(pointA, pointB)
print("Euclidean Distance:", euclidean_dist)

# Similarity (inverse of distance)
similarity_euclidean = 1 / (1 + euclidean_dist)
print("Euclidean Similarity:", similarity_euclidean)

```

```python
# Manhattan (City Block) Distance

import numpy as np
from scipy.spatial import distance

manhattan_dist = distance.cityblock(pointA, pointB)
print("Manhattan Distance:", manhattan_dist)

# Similarity (inverse of distance)
similarity_manhattan = 1 / (1 + manhattan_dist)
print("Manhattan Similarity:", similarity_manhattan)
```

```python
# Minkowski Distance (generalized form)

import numpy as np
from scipy.spatial import distance
minkowski_dist_p3 = distance.minkowski(pointA, pointB, p=3)
print("Minkowski Distance (p=3):", minkowski_dist_p3)

# Similarity (inverse of distance)
similarity_minkowski = 1 / (1 + minkowski_dist_p3)
print("Minkowski Similarity (p=3):", similarity_minkowski)

```

```python
# pearson Correlation
import pandas as pd
from scipy.stats import pearsonr

df = pd.DataFrame({
    'Hours':[1,2,3,4,5],
    'Marks':[20,35,50,65,80]
})

corr, p = pearsonr(df['Hours'], df['Marks'])

print("Pearson Correlation:", corr)
print("P-value:", p)
```

```python
# Spearman Correlation

import pandas as pd
from scipy.stats import spearmanr

df = pd.DataFrame({
    'X':[10,20,30,40,50],
    'Y':[12,18,33,47,55]
})

corr,p = spearmanr(df['X'], df['Y'])

print("Spearman Correlation:", corr)
print("P-value:", p)
```

```python
# 3. Edit-Based Measures (String/Text Data)
# Hamming Distance

def hamming_distance(str1, str2):
    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

s1 = "karolin"
s2 = "kathrin"

print("Hamming Distance:", hamming_distance(s1,s2))
```

```python
# Jaccard Index

def jaccard_index(str1, str2):
    set1 = set(str1.split())
    set2 = set(str2.split())

    intersection = set1.intersection(set2)
    union = set1.union(set2)

    return len(intersection)/len(union)

s1 = "data science is fun"
s2 = "science makes data useful"

print("Jaccard Index:", jaccard_index(s1,s2))
```

```python
# Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "data science is fun"
s2 = "science makes data useful"

vectorizer = CountVectorizer()

vectors = vectorizer.fit_transform([s1,s2])

similarity = cosine_similarity(vectors[0], vectors[1])

print("Cosine Similarity:", similarity[0][0])
```

```python
# Longest Common Subsequence (LCS)

def lcs_length(X, Y):

    m = len(X)
    n = len(Y)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                dp[i+1][j+1] = dp[i][j] + 1
            else:
                dp[i+1][j+1] = max(dp[i][j+1], dp[i+1][j])

    return dp[m][n]

seq1 = "ABCDEF"
seq2 = "AEBDF"

print("LCS Length:", lcs_length(seq1, seq2))
```
