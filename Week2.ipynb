```python
# A) Handling Missing Values

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

print("Original Data")
print(df)

# Mean Imputation
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Mode Imputation
df['Department'].fillna(df['Department'].mode()[0], inplace=True)

print("\nAfter Mean & Mode Imputation")
print(df)

```

```python
# Forward Fill

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

df.ffill(inplace=True)

print(df)
```

```python
# Backward Fill

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

df.bfill(inplace=True)

print(df)
```

```python
# Deletion

import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age':[25,30,np.nan,40,35],
    'Department':['HR','Finance','Finance',np.nan,'IT']
})

print("Drop Rows")
print(df.dropna())

print("\nDrop Columns")
print(df.dropna(axis=1))

```

```python
# B) Removing Duplicates

import pandas as pd

df = pd.DataFrame({
    'ID':[1,2,2,3,4,4],
    'Name':['Alice','Bob','Bob','Charlie','David','David'],
    'Age':[25,30,30,35,40,40]
})

print("Original Data")
print(df)

# Exact Match Removal
print("\nExact Duplicates Removed")
print(df.drop_duplicates())

# Subset-Based Removal
print("\nDuplicates Removed Based on ID")
print(df.drop_duplicates(subset=['ID']))

```

```python
# C) Correcting Inconsistent Formats
# Date Standardization

import pandas as pd

df = pd.DataFrame({
    'Date':['2025-01-05','05/01/2025','Jan 5, 2025','2025.01.05']
})

df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
# Case Normalization

import pandas as pd

df = pd.DataFrame({
    'Name':['Alice','BOB','charlie','DAVID']
})

df['Lower'] = df['Name'].str.lower()
df['Upper'] = df['Name'].str.upper()

print(df)
```
